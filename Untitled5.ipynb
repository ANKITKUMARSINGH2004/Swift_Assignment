## Code
Below are the codes that demonstrates analyzing the data using Python:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load data

```python
sales_data = pd.read_csv("C:\\Users\\ankit\\Downloads\\Sales_Data.csv")
glance_views = pd.read_csv("C:\\Users\\ankit\\Downloads\\Glance_Views.csv")
```

```python
sales_data.head()
```

```python
glance_views.head()
```

## Data Overview]


```python
print("\nSales Data Summary:")
print(sales_data.describe())
print("\nGlance Views Summary:")
print(glance_views.describe())
```

## Check For Missing Values

```python
print("\nMissing Values in Sales Data:")
print(sales_data.isnull().sum())
print("\nMissing Values in Glance Views:")
print(glance_views.isnull().sum())
```

## Fill or Drop missing values

```python
sales_data = sales_data.dropna()  # Example: dropping missing values
glance_views = glance_views.dropna()
```

## Data Merging based on SKU_NAME and FEED_DATE

```python
merged_data = pd.merge(sales_data, glance_views, on=["SKU_NAME", "FEED_DATE"])
print("\nMerged Data Overview:")
print(merged_data.head())
```

## Adding calculated fields

```python
merged_data['Conversion_Rate'] = merged_data['UNITS'] / merged_data['VIEWS']
merged_data['Revenue_Per_Unit'] = merged_data['ORDERED_REVENUE'] / merged_data['ORDERED_UNITS']
```

## EDA Visualizations
## 1. Revenue Distributions

```python
plt.figure(figsize=(8, 6))
sns.histplot(sales_data['ORDERED_REVENUE'], kde=True)
plt.title("Revenue Distribution")
plt.xlabel("Ordered Revenue")
plt.ylabel("Frequency")
plt.show()
```

## 2. Conversion Rate vs Revenue

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=merged_data, x='Conversion_Rate', y='ORDERED_REVENUE')
plt.title("Conversion Rate vs Ordered Revenue")
plt.xlabel("Conversion Rate")
plt.ylabel("Ordered Revenue")
plt.show()
```

## 3. Category-wise Revenue

```python
category_revenue = sales_data.groupby("SUB_CATEGORY")["ORDERED_REVENUE"].sum().sort_values()
plt.figure(figsize=(8, 6))
category_revenue.plot(kind="bar", color="skyblue")
plt.title("Category-wise Ordered Revenue")
plt.xlabel("Category")
plt.ylabel("Total Ordered Revenue")
plt.show()
```

## Correlation Analytics

```python
correlation_matrix = merged_data[['ORDERED_REVENUE', 'ORDERED_UNITS', 'VIEWS', 'UNITS', 'Conversion_Rate']].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## Exporting Results to CSV

```python
merged_data.to_csv("eda_results.csv", index=False)
print("\nEDA completed and results saved to 'eda_results.csv'")
```
